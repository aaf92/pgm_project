```python
# load data
prot_df = pd.read_csv('/ix/djishnu/Aaron_F/PGM_project/TLR_data/Results/reduced_proteomics_adj.pval_0.05_varThresh_0.80.csv')
print(prot_df.shape)
```

```python
# get correlation matrix
prot_df.set_index(prot_df.columns[0], inplace=True)
avg_prot_df = prot_df.groupby(lambda x: x.split('_', 1)[1], axis=1).mean()
```

```python
# calc coexpression matrix
coexpression_matrix = avg_prot_df.T.corr().abs()
```

```python
# check for nas
print(coexpression_matrix.isna().sum().sum())
```

```python
coexpression_matrix
```

```python
# make p-sites unique by taking the max_coexpression average value between psites mapped to the same gene
psite_map_df = pd.read_csv("/ix/djishnu/Aaron_F/PGM_project/TLR_data/TLR_psites_mapped.csv")
psites = prot_df.index.tolist()
psite_map_df = psite_map_df.loc[psite_map_df["id"].isin(psites),]
psite_gene_dict = psite_map_df.groupby('geneSymbol')['id'].apply(list).to_dict()
psite_gene_dict
```

```python
psites_to_keep = []
# go through each gene and if multiple p-sites then find max
for key, val in psite_gene_dict.items():
    if len(val) > 1:
        coexp_means = coexpression_matrix[val].mean()
        max_mean_col = coexp_means.idxmax()
        psites_to_keep.append(max_mean_col)
    else:
        psites_to_keep.append(val[0])
```

```python
avg_prot_df.reset_index(inplace=True)
```

```python
# subset df to only contain psites that are unique and have highest coexpression value
avg_prot_df = avg_prot_df.loc[avg_prot_df["id"].isin(psites_to_keep),]
```

```python
avg_prot_df = pd.merge(avg_prot_df, psite_map_df, how="left", on=["id"])
```

```python
avg_prot_df = avg_prot_df.set_index('geneSymbol').drop(columns=['id'])
```

```python
final_coexp_df = avg_prot_df.T.corr().abs()
```

```python
# create sparsity for Bionic by raising values to power of six
final_coexp_df = final_coexp_df ** 6
```

```python
# save coexpression edge data
# Loop over the upper triangular part of the matrix (excluding the diagonal)
coexpression_values = []
for i in range(len(final_coexp_df.columns)):
    for j in range(i + 1, len(final_coexp_df.columns)):
        gene1 = final_coexp_df.index[i]
        gene2 = final_coexp_df.columns[j]
        exp_val = final_coexp_df.iat[i, j]  # Access the coexpression value
        coexpression_values.append(f"{gene1}\t{gene2}\t{exp_val}")

# Write the results to a text file
with open("TLR_prot_abs_coexpression_edge_vals_sparse.txt", "w") as f:
    for line in coexpression_values:
        f.write(line + "\n")
```
