### determine if phosphostires are up or down regulated based on 45 min time point

```python
TLR_data_rep1['state_R1'] = TLR_data_rep1.iloc[:, 2].apply(lambda x: 1 if x > 0 else -1)
TLR_data_rep2['state_R2'] = TLR_data_rep2.iloc[:, 2].apply(lambda x: 1 if x > 0 else -1)
```

```python
TLR_data_rep1
```

```python
TLR_data_rep2
```

```python
TLR_combined = pd.concat([TLR_data_rep1, TLR_data_rep2], axis=1)
```

```python
TLR_combined
```

```python
# Count the number of times state_R1 and state_R2 are similar
similar_count = (TLR_combined['state_R1'] == TLR_combined['state_R2']).sum()

# Count the number of times state_R1 and state_R2 are different
different_count = (TLR_combined['state_R1'] != TLR_combined['state_R2']).sum()

print(f"Number of similar states: {similar_count}")
print(f"Number of different states: {different_count}")
```

```python
TLR_combined
```

```python
# Create a new DataFrame TLR_states with the same index as TLR_combined
TLR_states = pd.DataFrame(index=TLR_combined.index)

# Iterate through each row in TLR_combined
for idx, row in TLR_combined.iterrows():
    if row['state_R1'] == row['state_R2']:
        TLR_states.loc[idx, 'final_state'] = row['state_R1']
    else:
        if (row['m1_KLA_WT_45m'] + row['m2_KLA_WT_45m']) / 2 > 1:
            TLR_states.loc[idx, 'final_state'] = int(1)
        else:
            TLR_states.loc[idx, 'final_state'] = int(-1)

print("TLR States DataFrame:")
print(TLR_states)
```

```python
TLR_states
```

```python
upreg = 0
downreg = 0

for i in TLR_states['final_state']:
    if i == 1:
        upreg += 1
        # print('Upregulated', upreg)
    else:
        downreg += 1
        # print('Downregulated', downreg)

print('Upregulated:', upreg)
print('Downregulated:', downreg)
```

```python
print('Count of +1 states:', upreg)
print('Count of -1 states:', downreg)
```

```python
TLR_states.to_csv('TLR_states.csv', index=True)
```
